```python
import pandas as pd
import numpy as np

file_path = "pollution_us_2000_2016.csv"

df = pd.read_csv(file_path, sep=",")
```

```python
df.head(10)
```

```python
df['SO2 AQI'].replace('', np.nan, inplace=True)
print (df)
```

```python
df['CO AQI'].replace('', np.nan, inplace=True)
print (df)
```

```python
df = df.dropna()
print (df)
```

```python
pollution_df = pd.DataFrame(df)
```

```python
pollution_df.drop_duplicates()
```

```python
pollution_df.isnull().count()
```

```python
pollution_df.head(25)
```

```python
pollution_AQI_cleaned = pollution_df[["State Code", "County Code", "Site Num", "Address", "State", "County", "City", "Date Local", "CO AQI"]]

pollution_AQI = pollution_AQI_cleaned.set_index("State Code")

print(pollution_AQI)
```

```python
pollution_AQI.describe()
```

```python
pollution_AQI["CO AQI"] = pollution_AQI["CO AQI"].astype(int64)
pollution_AQI.dtypes
```

```python
pollution_AQI.to_csv('Pollution_US_cleaned.csv')
```
